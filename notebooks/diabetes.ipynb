#train test

```python
from sklearn.model_selection import train_test_split #need 2 param data and label
```

```python
x=df.iloc[:,df.columns!='Outcome'] #data
y=df.iloc[:,df.columns=='Outcome']#outcome or label
print(y)
```

```python
xtrain,xtest,ytrain,ytest=train_test_split(x,y,test_size=0.2) #20%test,80%train
```

```python
xtrain.head()
```

```python
ytrain.head()
```

```python
xtrain.shape
ytrain.shape
```

ALGORTITHM


```python
from sklearn.ensemble import RandomForestClassifier
```

```python
model=RandomForestClassifier()
```

```python
model.fit(xtrain,ytrain.values.ravel()) # to train alg
```

```python
predict_output=model.predict(xtest) #to test alg
print(predict_output)
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(predict_output,ytest)
```
